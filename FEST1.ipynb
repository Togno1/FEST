```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#PUNTO 7
```

```python
a = 1.0        
N = 1000       
M = 10000     
```

```python
steps = np.random.choice([-a, a], size=(M, N))
positions = steps.sum(axis=1)
```

```python
plt.hist(positions, bins=30, density=True, alpha=0.6, label="Simulación")

mu = 0
sigma = np.sqrt(N*a**2)
x_vals = np.linspace(positions.min(), positions.max(), 200)
plt.plot(x_vals, norm.pdf(x_vals, mu, sigma), 'r-', lw=2, label="Gaussiana")

plt.xlabel("Posición final x")
plt.ylabel("Probabilidad")
plt.title(f"Distribución de posiciones finales después de N={N} pasos")
plt.legend()
plt.show()
```

```python
N_values = [50, 100, 200, 500, 1000]
mean_x2 = []

for N in N_values:
    steps = np.random.choice([-a, a], size=(M, N))
    positions = steps.sum(axis=1)
    mean_x2.append(np.mean(positions**2))
```

```python
plt.plot(N_values, mean_x2, 'o-', label="Simulación")
plt.plot(N_values, [a**2*N for N in N_values], 'r--', label="Teoría")

plt.xlabel("N")
plt.ylabel("<x^2>")
plt.title("<x^2> en función de N")
plt.legend()
plt.show()
```
